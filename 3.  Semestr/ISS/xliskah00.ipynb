```python
# load the data
# references will be in ref_signals, reference labels in ref_labels, reference count in N_ref.
# tests  will be in test_signals, test labels in test_labels, test count in N_test.
def get_signals(labs):
    signals = []
    N = len(labs)
    for car in labs:
      filename = login + "/" + car + ".wav"
      s, Fs = sf.read(filename)
      signals.append(s)
    return signals, N, Fs

def play_signals(signals, Fs):
  for signal in signals:
    display(Audio(signal, rate=Fs))

files = glob.glob(login + "/*.wav")
names = [re.sub(login + "/", "", s) for s in files]
labels = [re.sub(".wav", "",s) for s in names]
print ("----- test signals ---------")
r = re.compile("^test_");  test_labels = list(filter(r.match, labels))
print (test_labels); test_signals, N_test, Fs = get_signals(test_labels); play_signals (test_signals, Fs)
print ("----- reference signals ---------")
r = re.compile("(?!^test_)");  ref_labels = list(filter(r.match, labels))
print (ref_labels); ref_signals, N_ref, Fs = get_signals(ref_labels); play_signals (ref_signals, Fs)
```

```python


def plot_signals(signals, labels):
    """
    Plot the signals with appropriate labels for reference and test signals.
    The first few samples of the signal are displayed for visualization.
    """
    for i, signal in enumerate(signals):
        plt.figure(figsize=(15, 4))

        # Plot only the first part of the signal for visualization purposes
        if (i < 3 and labels[i] in ref_labels) or (i >= len(signals) - 2 and labels[i] in test_labels):
            plt.plot(signal[:4000])  # Plot only the first 4000 samples
            plt.title(f"Signal: {labels[i]} (First 4000 Samples)")
        elif (i < 2 and labels[i] in test_labels):
            plt.plot(signal[:8000])  # Plot the first 8000 samples for longer signals
            plt.title(f"Signal: {labels[i]} (First 8000 Samples)")
        else:
            plt.plot(signal)
            plt.title(f"Signal: {labels[i]}")

        plt.xlabel("Sample")
        plt.ylabel("Amplitude")
        plt.grid(True)
        plt.show()

# Display reference and test signals
print("Reference Signals:")
plot_signals(ref_signals, ref_labels)

print("\nTest Signals:")
plot_signals(test_signals, test_labels)

# ------------- FEATURE EXTRACTION (MFCC) -------------

def extract_mfcc_features(signals, Fs, n_mfcc=40):
    """
    Extract MFCC features from each signal.
    Combines MFCCs, deltas, and delta-deltas into a single feature vector.
    """
    mfcc_features = []
    for signal in signals:
        mfcc = librosa.feature.mfcc(y=signal, sr=Fs, n_mfcc=n_mfcc, hop_length=128, n_fft=2048)
        mfcc_delta = librosa.feature.delta(mfcc)
        mfcc_delta2 = librosa.feature.delta(mfcc, order=2)
        # Calculate mean and standard deviation for all coefficients
        mfcc_mean = np.mean(mfcc, axis=1)
        mfcc_std = np.std(mfcc, axis=1)
        mfcc_combined = np.concatenate((mfcc_mean, mfcc_std, np.mean(mfcc_delta, axis=1), np.std(mfcc_delta, axis=1), np.mean(mfcc_delta2, axis=1), np.std(mfcc_delta2, axis=1)))
        mfcc_features.append(mfcc_combined)
    return np.array(mfcc_features)

print("Extracting MFCC features for reference signals...")
ref_mfcc_features = extract_mfcc_features(ref_signals, Fs)

print("Extracting MFCC features for test signals...")
test_mfcc_features = extract_mfcc_features(test_signals, Fs)

# ------------- DISTANCE AND CORRELATION CALCULATION -------------

def calculate_distances(test_features, ref_features):
    """ Calculate the Euclidean distances between test features and reference features. """
    distances = euclidean_distances(test_features, ref_features)
    return distances

def calculate_correlation_coefficients(test_features, ref_features):
    """ Calculate the Pearson correlation coefficients between test features and reference features. """
    correlations = np.zeros((len(test_features), len(ref_features)))
    for i, test_feat in enumerate(test_features):
        for j, ref_feat in enumerate(ref_features):
            test_feat_normalized = (test_feat - np.mean(test_feat)) / np.std(test_feat)
            ref_feat_normalized = (ref_feat - np.mean(ref_feat)) / np.std(ref_feat)
            corr, _ = pearsonr(test_feat_normalized, ref_feat_normalized)
            correlations[i, j] = corr
    return correlations

# Calculate distances and correlations
print("Calculating distances and correlations...")
distances = calculate_distances(test_mfcc_features, ref_mfcc_features)
correlations = calculate_correlation_coefficients(test_mfcc_features, ref_mfcc_features)

# ------------- CLASSIFICATION -------------

def classify_signals(distances, correlations, ref_labels):
    """
    Classify each test signal by comparing it to reference signals based on distance and correlation.
    Identifies definite matches and test signals that do not match any reference.
    """
    predicted_labels = []
    assigned_references = set()
    unassigned_test_signals = set(range(distances.shape[0]))
    definite_matches = []
    no_matches = []

    for i in range(distances.shape[0]):
        for j in range(distances.shape[1]):
            print(f"Test signal {i} to Reference {ref_labels[j]}: Distance = {distances[i, j]:.4f}, Correlation = {correlations[i, j]:.4f}")
        min_index_dist = np.argmin(distances[i, :])
        max_index_corr = np.argmax(correlations[i, :])
        if distances[i, min_index_dist] > 200 or correlations[i, max_index_corr] < 0.95:
            print(f"NO MATCH: Test signal '{test_labels[i]}' has no match\n")
            predicted_labels.append(("No match", "No match"))
            no_matches.append(test_labels[i])
        else:
            if min_index_dist == max_index_corr and ref_labels[min_index_dist] not in assigned_references:
                print(f"DEFINITE MATCH: Test signal '{test_labels[i]}' is definitely matched to reference '{ref_labels[min_index_dist]}'\n")
                assigned_references.add(ref_labels[min_index_dist])
                unassigned_test_signals.discard(i)
                definite_matches.append((test_labels[i], ref_labels[min_index_dist]))
            predicted_labels.append((ref_labels[min_index_dist], ref_labels[max_index_corr]))

    while unassigned_test_signals:
        best_score = float('inf')
        best_test_idx = None
        best_ref_idx = None
        for i in unassigned_test_signals:
            for j in range(distances.shape[1]):
                if ref_labels[j] not in assigned_references:
                    score = distances[i, j] - correlations[i, j]
                    if score < best_score:
                        best_score = score
                        best_test_idx = i
                        best_ref_idx = j
        if best_test_idx is not None and best_ref_idx is not None:
            if distances[best_test_idx, best_ref_idx] <= 200 and correlations[best_test_idx, best_ref_idx] >= 0.95:
                print(f"DEFINITE MATCH: Test signal '{test_labels[best_test_idx]}' is matched to reference '{ref_labels[best_ref_idx]}'\n")
                assigned_references.add(ref_labels[best_ref_idx])
                unassigned_test_signals.discard(best_test_idx)
                definite_matches.append((test_labels[best_test_idx], ref_labels[best_ref_idx]))
            else:
                print(f"NO MATCH: Test signal '{test_labels[best_test_idx]}' has no match\n")
                no_matches.append(test_labels[best_test_idx])
                unassigned_test_signals.discard(best_test_idx)

    return predicted_labels, definite_matches, no_matches

# Classify signals
predicted_labels, definite_matches, no_matches = classify_signals(distances, correlations, ref_labels)

# Print classification results
print("\n----- CLASSIFICATION RESULTS -----")
for i, label in enumerate(test_labels):
    print(f"Test signal '{label}' is most similar to reference signal '{predicted_labels[i][0]}' based on distance and '{predicted_labels[i][1]}' based on correlation.")

# Print definite matches
print("\n----- DEFINITE MATCHES -----")
for match in definite_matches:
    print(f"Test signal '{match[0]}' is definitely matched to reference '{match[1]}'")

# Print no matches
for no_match in no_matches:
    print(f"Test signal '{no_match}' has no match")



# Visualization of distance and correlation heatmaps
def visualize_heatmap(matrix, title, x_labels, y_labels, cmap, vmin=None, vmax=None):
    plt.figure(figsize=(14, 12))
    plt.imshow(matrix, interpolation='nearest', cmap=cmap, vmin=vmin, vmax=vmax)
    plt.colorbar(label=title)
    plt.xlabel('Reference Signals')
    plt.ylabel('Test Signals')
    plt.xticks(ticks=np.arange(len(x_labels)), labels=x_labels, rotation=45)
    plt.yticks(ticks=np.arange(len(y_labels)), labels=y_labels)
    plt.title(title)
    plt.show()

visualize_heatmap(distances, 'Distance Heatmap', ref_labels, test_labels, cmap='viridis')
visualize_heatmap(correlations, 'Correlation Heatmap', ref_labels, test_labels, cmap='coolwarm', vmin=0.8, vmax=1)

```
